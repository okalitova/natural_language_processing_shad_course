Ğ£ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ğ° Ğ² Unicode ĞµÑÑ‚ÑŒ ÑĞ²Ğ¾Ğ¹ÑÑ‚Ğ²Ğ° (property), ĞºĞ¾Ñ‚Ğ¾Ñ€Ñ‹Ğµ Ğ¼Ğ¾Ğ³ÑƒÑ‚ ÑĞ¾Ğ¾Ğ±Ñ‰Ğ¸Ñ‚ÑŒ Ğ½Ğ°Ğ¼ Ğ¼Ğ½Ğ¾Ğ³Ğ¾ Ğ¿Ğ¾Ğ»ĞµĞ·Ğ½Ğ¾Ğ¹ Ğ¸Ğ½Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ†Ğ¸Ğ¸ Ğ¾ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ğµ. ĞŸĞ¾Ğ¿Ñ€Ğ¾Ğ±ÑƒĞµĞ¼ Ñ Ğ½Ğ¸Ğ¼Ğ¸ Ğ½ĞµĞ¼Ğ½Ğ¾Ğ³Ğ¾ Ğ¿Ğ¾Ğ¸Ğ³Ñ€Ğ°Ñ‚ÑŒÑÑ.

1. ĞĞ¿Ñ€ĞµĞ´ĞµĞ»Ğ¸Ñ‚Ğµ, Ğº ĞºĞ°ĞºĞ¾Ğ¹ ĞºĞ°Ñ‚ĞµĞ³Ğ¾Ñ€Ğ¸Ğ¸ Ğ¾Ñ‚Ğ½Ğ¾ÑÑÑ‚ÑÑ Ñ†Ğ¸Ñ„Ñ€Ñ‹ Ğ² Ñ€Ğ°Ğ·Ğ»Ğ¸Ñ‡Ğ½Ñ‹Ñ… Ğ¿Ğ¸ÑÑŒĞ¼ĞµĞ½Ğ½Ğ¾ÑÑ‚ÑÑ….
2. ĞĞ°Ğ¹Ğ´Ğ¸Ñ‚Ğµ Ğ¸Ñ… Ğ²ÑĞµ, ÑĞ¾Ñ…Ñ€Ğ°Ğ½Ğ¸Ñ‚Ğµ Ğ² ÑĞ»Ğ¾Ğ²Ğ°Ñ€ÑŒ {character: value} Ğ¸ Ğ²Ñ‹Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ğ½Ğ° ÑĞºÑ€Ğ°Ğ½.
3. ĞĞ°Ğ¿Ğ¸ÑˆĞ¸Ñ‚Ğµ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ, Ñ€Ğ°ÑĞ¿Ğ¾Ğ·Ğ½Ğ°ÑÑ‰ÑƒÑ Ñ‡Ğ¸ÑĞ»Ğ°, ÑĞ¾ÑÑ‚Ğ¾ÑÑ‰Ğ¸Ğµ Ğ¸Ğ· Ñ†Ğ¸Ñ„Ñ€ Ñ€Ğ°Ğ·Ğ»Ğ¸Ñ‡Ğ½Ñ‹Ñ… Ğ¿Ğ¸ÑÑŒĞ¼ĞµĞ½Ğ½Ğ¾ÑÑ‚ĞµĞ¹, Ğ¸ Ñ€Ğ°ÑĞ¿Ğ¾Ğ·Ğ½Ğ°Ğ¹Ñ‚Ğµ **8Ù¥Ûµà¥¯à¥­à©©à«§**.

```python
import unicodedata

mapping = {}
unicode_chars_number = 1114112
for char in range(0, unicode_chars_number):
    try:
        if (unicodedata.category(chr(char)) == "Nd"):
            mapping[chr(char)] = unicodedata.digit(chr(char))
    except ValueError:
        pass
print(mapping)

def decode_digits(string, mapping):
    digits = []
    for char in string:
        digits.append(mapping[char])
    return ''.join(map(str, digits))

print(decode_digits(u'8Ù¥Ûµà¥¯à¥­à©©à«§', mapping))
```

ĞŸĞ¾Ñ€Ğ°Ğ±Ğ¾Ñ‚Ğ°ĞµĞ¼ Ñ Ğ°Ñ€Ğ°Ğ±ÑĞºĞ¸Ğ¼ Ğ°Ğ»Ñ„Ğ°Ğ²Ğ¸Ñ‚Ğ¾Ğ¼. Ğ’Ğ¾Ğ·Ğ¼Ğ¾Ğ¶Ğ½Ğ¾, Ğ´Ğ»Ñ ÑÑ‚Ğ¾Ğ³Ğ¾ Ğ²Ğ°Ğ¼ Ğ¿Ğ¾Ğ½Ğ°Ğ´Ğ¾Ğ±Ğ¸Ñ‚ÑÑ Ğ·Ğ½Ğ°Ñ‚ÑŒ Ñ…Ğ¾Ñ‚Ñ Ğ±Ñ‹ Ğ¾Ğ´Ğ½Ñƒ Ğ°Ñ€Ğ°Ğ±ÑĞºÑƒÑ Ğ±ÑƒĞºĞ²Ñƒ. Ğš Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ñƒ, Ğ²Ğ¾Ğ·ÑŒĞ¼ĞµĞ¼ ÑĞ»ĞµĞ´ÑƒÑÑ‰ÑƒÑ **Ø¹**.

1. ĞĞ°Ğ¿Ğ¸ÑˆĞ¸Ñ‚Ğµ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ, Ğ½Ğ°Ñ…Ğ¾Ğ´ÑÑ‰ÑƒÑ Ğ²ÑĞµ Ğ°Ñ€Ğ°Ğ±ÑĞºĞ¸Ğµ Ğ±ÑƒĞºĞ²Ñ‹ Ñ Ğ¾Ğ³Ğ»Ğ°ÑĞ¾Ğ²ĞºĞ°Ğ¼Ğ¸ Ğ¸ Ğ±ĞµĞ·.  
_ĞŸĞ¾Ğ´ÑĞºĞ°Ğ·ĞºĞ°_: Ğ²Ğ¾ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞ¹Ñ‚ĞµÑÑŒ Ñ„ÑƒĞ½ĞºÑ†Ğ¸ĞµĞ¹ name.
2. ĞĞ°Ğ¿Ğ¸ÑˆĞ¸Ñ‚Ğµ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ, Ğ½Ğ°Ñ…Ğ¾Ğ´ÑÑ‰ÑƒÑ Ğ²ÑĞµ Ğ°Ñ€Ğ°Ğ±ÑĞºĞ¸Ğµ Ğ±ÑƒĞºĞ²Ñ‹ Ğ±ĞµĞ· Ğ¾Ğ³Ğ»Ğ°ÑĞ¾Ğ²Ğ¾Ğº (Ğ¾Ğ±Ñ‹Ñ‡Ğ½Ğ¾ Ğ² name, ĞºĞ¾Ğ³Ğ´Ğ° Ğ±ÑƒĞºĞ²Ğ° Ñ Ğ¾Ğ³Ğ»Ğ°ÑĞ¾Ğ²ĞºĞ¾Ğ¹, Ğ¿Ğ¸ÑˆÑƒÑ‚ "WITH ...").

```python
def get_arabic_letters():
    arabic_letters = []
    unicode_chars_number = 1114112
    for char in range(0, unicode_chars_number):
        try:
            if unicodedata.name(chr(char)).find("ARABIC LETTER") != -1:
                arabic_letters.append(chr(char))
        except ValueError:
            pass
    return arabic_letters

def get_clean_arabic_letters():
    clean_arabic_letters = []
    for arrabic_letter in get_arabic_letters():
        if unicodedata.name(arrabic_letter).find("WITH") == -1:
            clean_arabic_letters.append(arrabic_letter)
    return clean_arabic_letters

print("arabic letters")
print(''.join(get_arabic_letters()))
print("arabic letters without ...")
print(''.join(get_clean_arabic_letters()))
```

ĞŸĞ¾Ğ¿Ñ€Ğ¾Ğ±ÑƒĞ¹Ñ‚Ğµ Ğ½Ğ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ¾Ğ²Ğ°Ñ‚ÑŒ Ğ°Ñ€Ğ°Ğ±ÑĞºĞ¸Ğµ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ñ‹ Ğ¿Ñ€Ğ¸ Ğ¿Ğ¾Ğ¼Ğ¾Ñ‰ÑŒÑ Ñ€Ğ°Ğ·Ğ»Ğ¸Ñ‡Ğ½Ñ‹Ñ… Ğ½Ğ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ°Ñ‚Ğ¾Ñ€Ğ¾Ğ² Ğ¸ Ğ¿Ñ€Ğ¾Ğ²ĞµÑ€ÑŒÑ‚Ğµ, ÑĞºĞ»ĞµÑÑ‚ÑÑ Ğ»Ğ¸ ĞºĞ°ĞºĞ¸Ğµ-Ğ½Ğ¸Ğ±ÑƒĞ´ÑŒ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»Ñ‹, Ğ¸Ğ»Ğ¸ Ğ½ĞµÑ‚?

```python
def get_same_normalized_arabic_letters(form):
    arabic_letters = get_arabic_letters()
    normalized_arabic_letters = set()
    same_normalized_arabic_letters = set()
    for arabic_letter in arabic_letters:
        normalized_arabic_letter = unicodedata.normalize(form, arabic_letter)
        if normalized_arabic_letter in normalized_arabic_letters:
            same_normalized_arabic_letters.add(normalized_arabic_letter)
        else:
            normalized_arabic_letters.add(normalized_arabic_letter)
    return same_normalized_arabic_letters

print(''.join(get_same_normalized_arabic_letters('NFC')))
print(''.join(get_same_normalized_arabic_letters('NFKC')))
print(''.join(get_same_normalized_arabic_letters('NFD')))
print(''.join(get_same_normalized_arabic_letters('NFKD')))
```

ĞĞ°Ğ¿Ğ¸ÑˆĞ¸Ñ‚Ğµ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ¿Ğ¾ ÑƒĞ´Ğ°Ğ»ĞµĞ½Ğ¸Ñ Ğ´Ğ¸Ğ°ĞºÑ€Ğ¸Ñ‚Ğ¸ĞºĞ¸ Ğ¸Ğ· Ñ‚ĞµĞºÑÑ‚Ğ°.

```python
sent = u'ĞšĞ»Ñ‘Ğ²Ñ‹Ğ¹ Ğ¹ĞµÑ‚Ñ‚Ğ¸ Ñ€Ğ°Ğ·Ğ¶Ñ‘Ğ³ ĞºĞ¾ÑÑ‚Ñ‘Ñ€ Ğ¸Ğ· Ğ±ĞµÑ€Ñ‘Ğ·Ñ‹, Ñ‘Ğ»ĞºĞ¸ Ğ¸ ĞºĞ»Ñ‘Ğ½Ğ°.'

def delete_diacritic(sent):
    return ''.join(c for c in unicodedata.normalize('NFKD', sent)
                  if unicodedata.category(c) != 'Mn')

print(delete_diacritic(sent))
```
