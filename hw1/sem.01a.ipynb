## Unicode

У каждого символа в Unicode есть свойства (property), которые могут сообщить нам много полезной информации о символе. Попробуем с ними немного поиграться.

1. Определите, к какой категории относятся цифры в различных письменностях.
2. Найдите их все, сохраните в словарь {character: value} и выведите на экран.
3. Напишите функцию, распознающую числа, состоящие из цифр различных письменностей, и распознайте **8٥۵९७੩૧**.

```python
import unicodedata

mapping = {}
unicode_chars_number = 1114112
for char in range(0, unicode_chars_number):
    try:
        if (unicodedata.category(chr(char)) == "Nd"):
            mapping[chr(char)] = unicodedata.digit(chr(char))
    except ValueError:
        pass
print(mapping)

def decode_digits(string, mapping):
    digits = []
    for char in string:
        digits.append(mapping[char])
    return ''.join(map(str, digits))

print(decode_digits(u'8٥۵९७੩૧', mapping))
```

Поработаем с арабским алфавитом. Возможно, для этого вам понадобится знать хотя бы одну арабскую букву. К примеру, возьмем следующую **ع**.

1. Напишите функцию, находящую все арабские буквы с огласовками и без.  
_Подсказка_: воспользуйтесь функцией name.
2. Напишите функцию, находящую все арабские буквы без огласовок (обычно в name, когда буква с огласовкой, пишут "WITH ...").

```python
def get_arabic_letters():
    arabic_letters = []
    unicode_chars_number = 1114112
    for char in range(0, unicode_chars_number):
        try:
            if unicodedata.name(chr(char)).find("ARABIC LETTER") != -1:
                arabic_letters.append(chr(char))
        except ValueError:
            pass
    return arabic_letters

def get_clean_arabic_letters():
    clean_arabic_letters = []
    for arrabic_letter in get_arabic_letters():
        if unicodedata.name(arrabic_letter).find("WITH") == -1:
            clean_arabic_letters.append(arrabic_letter)
    return clean_arabic_letters

print("arabic letters")
print(''.join(get_arabic_letters()))
print("arabic letters without ...")
print(''.join(get_clean_arabic_letters()))
```

Попробуйте нормализовать арабские символы при помощью различных нормализаторов и проверьте, склеются ли какие-нибудь символы, или нет?

```python
def get_same_normalized_arabic_letters(form):
    arabic_letters = get_arabic_letters()
    normalized_arabic_letters = set()
    same_normalized_arabic_letters = set()
    for arabic_letter in arabic_letters:
        normalized_arabic_letter = unicodedata.normalize(form, arabic_letter)
        if normalized_arabic_letter in normalized_arabic_letters:
            same_normalized_arabic_letters.add(normalized_arabic_letter)
        else:
            normalized_arabic_letters.add(normalized_arabic_letter)
    return same_normalized_arabic_letters

print(''.join(get_same_normalized_arabic_letters('NFC')))
print(''.join(get_same_normalized_arabic_letters('NFKC')))
print(''.join(get_same_normalized_arabic_letters('NFD')))
print(''.join(get_same_normalized_arabic_letters('NFKD')))
```

Напишите функцию по удалению диакритики из текста.

```python
sent = u'Клёвый йетти разжёг костёр из берёзы, ёлки и клёна.'

def delete_diacritic(sent):
    return ''.join(c for c in unicodedata.normalize('NFKD', sent)
                  if unicodedata.category(c) != 'Mn')

print(delete_diacritic(sent))
```
